```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read the data from CSV
data = pd.read_csv('./comparative_analysis')
```

```python
# Converting the time data to a numerical format
data['Duration'] = pd.to_timedelta(data['Time needed for processing 100 selected sequences [h:m:s:ms]'])

# Creating the plot
plt.figure(figsize=(8, 6))
plt.scatter(data['Number of HPC threads'], data['Duration'].dt.total_seconds(), marker='o')
plt.xlabel('Number of HPC threads')
plt.ylabel('Time (seconds)')
plt.title('Processing Time vs. Number of HPC threads')
plt.grid(True)

#Saving the image 
plt.savefig("Comparative_analysis_graph", dpi=300, bbox_inches='tight')  

# Showing the plot
plt.show()
```
